```python
import numpy as np
import pandas as pd
```

## Load Datas

```python
train = pd.read_csv("./datas/train.csv", parse_dates=["datetime"])

print(train.shape)
train.head()
```

```python
# 위에서 parse_dates를 했기때문에 .dt.month를 사용할 수 있는 것임.
train["datetime"].dt.month.head()
```

```python
test = pd.read_csv("./datas/test.csv", parse_dates=["datetime"])

print(test.shape)
test.head()
```

```python
test["datetime"].dt.day.head()
```

## Train

```python
train["datetime-year"] = train["datetime"].dt.year
train["datetime-month"] = train["datetime"].dt.month
train["datetime-day"] = train["datetime"].dt.day
train["datetime-hour"] = train["datetime"].dt.hour
train["datetime-minute"] = train["datetime"].dt.minute
train["datetime-second"] = train["datetime"].dt.second

print(train.shape)
train[["datetime", "datetime-year", "datetime-month", "datetime-day", "datetime-hour", "datetime-minute", "datetime-second"]].head()
```

```python
test["datetime-year"] = test["datetime"].dt.year
test["datetime-month"] = test["datetime"].dt.month
test["datetime-day"] = test["datetime"].dt.day
test["datetime-hour"] = test["datetime"].dt.hour
test["datetime-minute"] = test["datetime"].dt.minute
test["datetime-second"] = test["datetime"].dt.second

print(test.shape)
test[["datetime", "datetime-year", "datetime-month", "datetime-day", "datetime-hour", "datetime-minute", "datetime-second"]].head()
```

```python
train["datetime-dayofweek"] = train["datetime"].dt.dayofweek

print(train.shape)
train[["datetime", "datetime-dayofweek"]].head()
```

```python
test["datetime-dayofweek"] = test["datetime"].dt.dayofweek

print(test.shape)
test[["datetime", "datetime-dayofweek"]].head()
```

```python
train.loc[(train["datetime-hour"] >= 6) & (train["datetime-hour"] <=9), "commute-hour"] = 1
train.loc[(train["datetime-hour"] >= 17) & (train["datetime-hour"] <=20), "commute-hour"] = 1
train["commute-hour"] = (train["commute-hour"] == 1)

test.loc[(test["datetime-hour"] >= 6) & (test["datetime-hour"] <=9), "commute-hour"] = 1
test.loc[(test["datetime-hour"] >= 17) & (test["datetime-hour"] <=20), "commute-hour"] = 1
test["commute-hour"] = (test["commute-hour"] == 1)
```

```python
train[train["windspeed"] == 0]
```

```python
# "datetime_year", "datetime_month", "datetime_day", "datetime_hour", "datetime_minute", "datetime_second"
# datetime_day <-- 36
feature_names = ["holiday"
                 , "workingday"
                 , "weather"
                 , "humidity"
                 , "atemp"
                 , "datetime-year"
                 , "datetime-month"
                 , "datetime-hour"
                 , "commute-hour"]

feature_names
```

```python
X_train = train[feature_names]

print(X_train.shape)
X_train.head()
```

```python
X_test = test[feature_names]

print(X_test.shape)
X_test.head()
```

```python
label_name = "count"
```

```python
Y_train = train[label_name]

print(Y_train.shape)
Y_train.head()
```

```python
# Kaggle에 바로 제출하지 않고도 정확도를 예측을 하기 위해서임.
# scikit-learn => sklearn
# from sklearn.tree import DecisionTreeRegressor
# model = DecisionTreeRegressor(random_state=3)

from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators = 20, random_state=3)

model
```

## Score <-- 중요함!

```python
# cross_validation 패키지를 model_selection로 옮겼음. model_selection 사용하는게 좋음.
from sklearn.cross_validation import cross_val_score
# from sklearn.model_selection import cross_val_score

score = cross_val_score(model, X_train, Y_train, cv=50, scoring="neg_mean_absolute_error").mean()

# 0에 가까울 수록 좋고, 0에 멀수록 안 좋음!!
print("Score = {0:5f}".format(score))
# 결과가 음수가 나옴. 상관없음. 양수로 출력하고 싶으면 절대값을 하든, -1을 곱하든 하면 됨.

# DecisionTreeRegressor : -152정도
# RandomForestRegressor : -129정도
# 117
# datetime <--- 36
# temp <-- 36비슷
# wind speed 제거 < 35
```

```python
model.fit(X_train, Y_train)
```

```python
predictions = model.predict(X_test)

print(predictions.shape)
predictions[0:10]
```

# Submit

```python
submission = pd.read_csv("./datas/sampleSubmission.csv")

submission["count"] = predictions

print(submission.shape)
submission.head()
```

```python
# index=False, csv 파일에서 0번째 컬럼을 없애기 위함.
submission.to_csv("./submission/baseline-script05.csv", index=False)
```
