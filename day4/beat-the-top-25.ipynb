```python
import pandas as pd
```

```python
train = pd.read_csv("data/train.csv", index_col="PassengerId")

print(train.shape)
train.head()
```

# Explore

```python
## 가장 많이 사용하는 시각화 모듈이나 사용하기 불편한 것이 있어서 seaborn의 도움을 받음.
## seaborn에서 없거나 사용하기 불편한 건 matplotlib에서 찾아쓰고 서로 보완하면서 사용하면 될 듯.
## 시각화는 major 5개로 해결을 하고, 여기서 해결이 안되면 seaborn을 찾아서 보면됨.
%matplotlib inline
import seaborn as sns
```

## 1. countplot

```python
# hue 기준으로 데이터를 쪼갠 후, 시각화를 각각함.
# 그냥 count를 세우고, x에 대해서만 갯수를 셈.
sns.countplot(data=train, x="Embarked", hue="Survived")
```

```python
sns.countplot(data=train, x="Pclass", hue="Survived")
```

## 2. barplot

```python
# count를 세기는 하나, x를 y에 대해서 평균값을 구함.
# 검정색 선은 표준편차임.
sns.barplot(data=train, x="Pclass", y="Fare", hue="Survived")
```

## 3. Pointplot

```python
# x축이 서로 연관이 있을 때는 pointplot을 사용하는게 좋다. x축이 관련이 없으면 barplot. 
# ex) bike의 시간에 해당하는 거
sns.pointplot(data=train, x="Pclass", y="Fare", hue="Survived")
```

## 4. distplot

```python
sns.distplot(train["Fare"])
```

```python
sns.distplot(train["Fare"], hist=False)
```

```python
low_fare = train[train["Fare"] < 100]

print(low_fare.shape)
low_fare.head()
```

```python
sns.distplot(low_fare["Fare"], hist=False)
```

```python
# 숫자에 대한 전반적인 분포를 알고 싶을 때, distplot을 사용한다.
perish = low_fare[low_fare["Survived"] == 0]
survived = low_fare[low_fare["Survived"] == 1]

sns.distplot(perish["Fare"], hist=False, label="Perish")
sns.distplot(survived["Fare"], hist=False, label="Survived")
```

## 5. Implot

```python
sns.lmplot(data=train, x="Age", y="Fare")

# 아래에 선이 보이는데, regression line임. 안 중요함.
```

```python
sns.lmplot(data=train, x="Age", y="Fare", fit_reg=False)
```

```python
sns.lmplot(data=train, x="Age", y="Fare", fit_reg=False, hue="Survived")
```

```python
# low_fare에 대해서 살펴봄.
sns.lmplot(data=low_fare, x="Age", y="Fare", fit_reg=False, hue="Survived")
```

## 6. subplots

```python
import matplotlib.pyplot as plt

figure, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3 )
```

```python
# nrows, ncols를 몇개를 하느냐에 따라 다르게 나와서 다루기가 약간 까다로움
figure, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3 )
figure.set_size_inches(20, 5) # size 조정

sns.countplot(data=train, x="Embarked", hue="Survived", ax=ax1)
sns.countplot(data=train, x="Sex", hue="Survived", ax=ax2)
sns.countplot(data=train, x="Pclass", hue="Survived", ax=ax3)
```

```python
figure, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(nrows=2, ncols=3)
figure.set_size_inches(20, 10) # size 조정

sns.countplot(data=train, x="Embarked", hue="Survived", ax=ax1)
sns.countplot(data=train, x="Sex", hue="Survived", ax=ax2)
sns.countplot(data=train, x="Pclass", hue="Survived", ax=ax3)
sns.countplot(data=train, x="SibSp", hue="Survived", ax=ax4)
sns.countplot(data=train, x="Parch", hue="Survived", ax=ax5)
```

```python
figure, ax = plt.subplots(nrows=1, ncols=1)
figure.set_size_inches(18, 4)

sns.countplot(data=train, x="SibSp", hue="Survived")
```
