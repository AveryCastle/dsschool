```python
import pandas as pd
```

```python
train = pd.read_csv("data/train.csv", index_col="PassengerId")

print(train.shape)
train
```

```python
test = pd.read_csv("data/test.csv", index_col="PassengerId")

print(test.shape)
test
```

### Preprocessing

### Encoding Sex 

#### computer가 쉽게 데이터를 처리할 수 있도록 0, 1로 변환하여 처리하기로 함.

```python
train.loc[train["Sex"] == "male", "Sex"] = 0
train.loc[train["Sex"] == "female", "Sex"] = 1

print(train.shape)
train
```

```python
test.loc[test["Sex"] == "male", "Sex"] = 0
test.loc[test["Sex"] == "female", "Sex"] = 1

print(test.shape)
test
```

```python
test.loc[test["Fare"].isnull(), "Fare"] = train["Fare"].mean()

test[test["Fare"].isnull()]
```

# Encode Embarked

```python
train["Embarked_C"] = train["Embarked"] == "C"
train["Embarked_S"] = train["Embarked"] == "S"
train["Embarked_Q"] = train["Embarked"] == "Q"

train[["Embarked", "Embarked_C", "Embarked_S", "Embarked_Q"]].head()
```

```python
test["Embarked_C"] = test["Embarked"] == "C"
test["Embarked_S"] = test["Embarked"] == "S"
test["Embarked_Q"] = test["Embarked"] == "Q"

test[["Embarked", "Embarked_C", "Embarked_S", "Embarked_Q"]].head()
```

```python
test[test["Embarked"].isnull()]
```

```python
test.head()
```

```python
mean_age = train["Age"].mean()
mean_age
```

```python
train[train["Age"].isnull()]
```

```python
train.loc[train["Age"].isnull(), "Age"] = mean_age

print(train.shape)
train.head(10)
```

```python
test.loc[test["Age"].isnull(), "Age"] = mean_age

print(test.shape)
test.head(10)
```

# Band Age

```python
train["Age_Young"] = train["Age"] < 15
train["Age_Medium"] = (train["Age"] >= 15) & (train["Age"] < 35)
train["Age_High"] = train["Age"] >= 35

train[["Age", "Age_Young", "Age_Medium", "Age_High"]].head()
```

```python
test["Age_Young"] = test["Age"] < 15
test["Age_Medium"] = (test["Age"] >= 15) & (test["Age"] < 35)
test["Age_High"] = test["Age"] >= 35

print(test.shape)
test[["Age", "Age_Young", "Age_Medium", "Age_High"]].head()
```

### Train

```python
feature_names = ["Pclass", "Sex", "Fare", "Embarked_C", "Embarked_S", "Embarked_Q", "Age_Young", "Age_Medium", "Age_High"]

feature_names
```

```python
X_train = train[feature_names]

print(X_train.shape)
X_train.head()
```

```python
X_test = test[feature_names]

print(X_test.shape)
X_test.head()
```

```python
label_name = "Survived"

y_train = train[label_name]

print(y_train.shape)
y_train.head()
```

```python
# scikit-learn => sklearn
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth=5)
model
```

```python
# Warning: You should install graphviz before run these codes.
# 1) Download Graphviz - http://www.graphviz.org/`Download.php
# 2) install graphviz - 'pip install graphviz' from terminal or install graphviz from Anaconda Navigator.

from sklearn.tree import export_graphviz
import graphviz

export_graphviz(model,
                feature_names=feature_names,
                class_names=["Perish", "Survived"],
                out_file="decision-tree.dot")

with open("decision-tree.dot") as f:
    dot_graph = f.read()
    
# graphviz.Source(dot_graph)
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
print (predictions.shape)
predictions[0:10]
```

# Submit

```python
submission = pd.read_csv("data/gender_submission.csv", index_col="PassengerId")

submission["Survived"] = predictions

print(submission.shape)
submission.head()
```

```python
submission.to_csv("data/baseline-age_script.csv")
```
